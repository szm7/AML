```python
employment_mapping = {
    '>=7': 'Highly Experienced',
    '1<=X<4': 'Experienced',
    '4<=X<7': 'Moderately Experienced',
    'unemployed': 'unemployed',
    '<1': 'Beginner'
}

data['employment_category'] = data['employment'].map(employment_mapping)
```

```python
data.head()
```

```python
'no known savings', '<100', '500<=X<1000', '>=1000', '100<=X<500'
employment_mapping = {
    '>=1000': 'High Savings',
    '500<=X<1000': 'Moderate Savings',
    '100<=X<500': 'Average Savings',
    '<100': 'Below average Savings',
    'no known savings': 'No Savings'
}

data['Average_Credit_Balance_category'] = data['Average_Credit_Balance'].map(employment_mapping)
```

```python
max_value = data['current_balance'].max()
min_value = data['current_balance'].min()

print('Maximum value:', max_value)
print('Minimum value:', min_value)
```

```python
data['class'].replace(to_replace='good', value=1, inplace=True)
data['class'].replace(to_replace='bad',  value=0, inplace=True)
data.head()
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X = data[['employment_category','Average_Credit_Balance_category']]
Y = data['class']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```
