```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
pio.templates.default = "plotly_white"
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/Program_4_Dataset.csv")
```

```python
data.head()
```

```python
x=data[['Temperature']].values
y=data['Pressure'].values
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('X Points')
plt.ylabel('Y Points')
plt.scatter(x,y)
plt.show()
```

```python
x1train,x1test,y1train,y1test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x2train,x2test,y2train,y2test=train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x3train,x3test,y3train,y3test=train_test_split(x, y, test_size=0.44, random_state=42)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

poly1 = PolynomialFeatures(degree=2)
X_train_p = poly1.fit_transform(x1train)
X_test_p = poly1.transform(x1test)

p1=PolynomialFeatures(degree=2,include_bias=True)
xtrain_p1=p1.fit_transform(x1train)
xtest_p1=p1.fit_transform(x1test)

model1 = LinearRegression()
model1.fit(X_train_p, y1train)

y_pred = model1.predict(X_test_p)

print('For degree  2')
r2 = r2_score(y1test, y_pred)
print(f'R-squared: {r2}')
mse = mean_squared_error(y1test, y_pred)
print(f'Mean Squared Error: {mse}')
mae = mean_absolute_error(y1test, y_pred)
print(f'Mean Absolute Error: {mae}')
```

```python
plt.scatter(x1train,model1.predict(X_train_p))
plt.scatter(x1train,y1train)
```
